```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
import numpy as np
```

```python
url="https://www.coindesk.com/"
itemsfull=[]
p = requests.get(url)
soup=BeautifulSoup(p.content,'html.parser')
content = soup.find_all('div',class_='article-cardstyles__AcTitle-q1x8lc-1 bwXBTf')
heading = ["News"]
for item in content:
    items=[]
    news = item.find('a',class_ = 'headline')
    items.append(news.text)
    itemsfull.append(items)
```

```python
df = pd.DataFrame(itemsfull,columns=['News'])
df.head()
```

```python
from googletrans import Translator
translator = Translator()
for i in range(len(df["News"])):
    op_english = translator.translate(df["News"][i], dest = "en")
    op_tamil = translator.translate(df["News"][i], dest = "ta")
    op_telugu = translator.translate(df["News"][i], dest = "te")
    
    print(i,op_english.text)
    print(i, op_tamil.text)
    print(i, op_telugu.text)
    print('\n')
```
